```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
r = 0.95  # Radius
phi = np.pi / 3  # Angle in radians

# Frequency range [0, 2*pi]
omega = np.linspace(0, 2 * np.pi, 1000)

# Compute H(e^{jω})
numerator = 1
denominator = 1 - 2 * r * np.cos(phi) * np.exp(-1j * omega) + r**2 * np.exp(-2j * omega)
H_omega = numerator / denominator

# Compute magnitude of H(omega)
abs_H_omega = np.abs(H_omega)

# Plot the magnitude response
plt.figure(figsize=(10, 6))
plt.plot(omega, abs_H_omega, label=r"$|H(\omega)|$", color="b")
plt.title("Magnitude Response of H(ω)")
plt.xlabel("Frequency (ω)")
plt.ylabel(r"$|H(\omega)|$")
plt.grid(True)
plt.legend()
plt.show()
```
