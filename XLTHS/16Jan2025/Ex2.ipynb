```python
#Library for coding
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Define common usage:
N = 20 #100 samples using to gather the data.
n = np.arange(0, N)
```

```python
#Define the signal
#x(n) = 2sin(0.3*pi*n) + 3cos(0.4*pi*n)
x_n = 2 * np.sin(0.3 * np.pi * n) + 3 * np.cos(0.4 * np.pi * n)
```

```python
# Compute DTFT
omega = np.linspace(0, 2 * np.pi, 20)  # Frequency range [0, 2*pi]
X_omega = np.array([np.sum(x_n * np.exp(-1j * w * n)) for w in omega])  # DTFT


#Change to the frequency domain 1000 points and 20 points for the time domain
# Compute DTFT
#omega = np.linspace(0, 2 * np.pi, 1000)  # Frequency range [0, 2*pi]
#X_omega = np.array([np.sum(x_n * np.exp(-1j * w * n)) for w in omega]) 
```

```python
# Compute magnitude of DTFT
abs_X_omega = np.abs(X_omega)
```

```python
# Plot the magnitude of X(omega)
plt.plot(omega, abs_X_omega)
plt.title("Magnitude of DTFT |X(omega)|")
plt.xlabel("Frequency (Ï‰)")
plt.ylabel(r"$|X(\omega)|$")
plt.grid(True)
plt.legend()
plt.show()
```
