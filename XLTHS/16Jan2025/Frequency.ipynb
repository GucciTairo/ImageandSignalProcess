```python
import matplotlib.pyplot as plt
import numpy as np
import sounddevice as sd
from scipy.signal import residue, lfilter
```

```python
#Define common usage:
N = 200 #100 samples using to gather the data.
n = np.arange(0, N)
```

```python
sigma = 0.1  # Standard deviation of noise
```

```python
# Sampling rate (in Hz) for audio playback
sampling_rate = 1000  # Standard audio sampling rate

# Time vector corresponding to the signal
duration = len(n) / sampling_rate  # Duration in seconds
t = np.linspace(0, duration, len(n), endpoint=False)
```

```python

# Signal components
x = 3 * np.cos(0.4 * np.pi * n) + 2 * np.sin(0.7 * np.pi * n)
```

```python
# Add Gaussian noise
v = np.random.normal(0, sigma, len(n))
x_n = x + v
```

```python
# Normalize the signal to the range [-1, 1]
x_n = x_n / np.max(np.abs(x_n))
```

```python
# Plot the signal
plt.plot(n, x_n)
plt.title("Signal x(n) with Gaussian Noise for x(n) = 3cos(0.4πn) + 2sin(0.7πn) + v(n)")
plt.xlabel("n")
plt.ylabel("x(n)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Play the sound
sd.play(x_n, samplerate=sampling_rate)
sd.wait()  # Wait for the sound to finish playing
```
