```python
import numpy as np
import matplotlib.pyplot as plt

# Define the signal
n = np.arange(0, 20, 0.1)  # Discrete time samples
x_n = np.sin(0.5 * np.pi * n)  # Signal: sin(0.5 * pi * n)

# Compute DTFT
omega = np.linspace(-np.pi, np.pi, 1000)  # Frequency range [-pi, pi]
X_omega = np.array([np.sum(x_n * np.exp(-1j * w * n)) for w in omega])  # DTFT

# Compute magnitude of DTFT
abs_X_omega = np.abs(X_omega)

# Plot the spectrum
plt.figure(figsize=(10, 6))
plt.plot(omega, abs_X_omega, label=r"$|X(\omega)|$", color="b")
plt.title("Spectrum of $sin(0.5 \pi n)$")
plt.xlabel("Frequency (ω)")
plt.ylabel(r"$|X(\omega)|$")
plt.grid(True)
plt.legend()
plt.show()
```
