```python
#Library
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Unit sample sequence
# Define the impulse function (n)
def delta(n, shift=0):
    #np.where(condition, x, y) returns x if condition is True(1), and y is False(0)
    return np.where(n == shift, 1, 0)
#In this case, if n == shift: The output is 1(the impulse occurs here)
#Else the output is 0
```

```python
# Define the range of n values
N = 40 # Number of samples
n = np.arange(-50, 50)  # From n = -100 to n = 100
```

```python
# Define h(n) = (n) - (n-10)
h = delta(n) - delta(n, shift=10)
```

```python
# Plot the function
plt.stem(n, h)
plt.title("h(n) = (n) - (n-10)")
plt.xlabel("n")
plt.ylabel("h(n)")
plt.grid(True)
plt.show()
```



Unit step sequence

```python
#Unit step sequence
#Define the unit step function u(n) conditionally
x = np.zeros_like(n)  # Initialize with zeros
x[n >= 0] = 1        # Set values to 1 for n >= 0
```

```python
plt.stem(n, x)
plt.title("x(n) = u(n)")
plt.xlabel("n")
plt.ylabel("x(n)")
plt.grid(True)
plt.axis()
plt.show()
```

```python
#Unit step sequence for x(n) = u(n-10)
#Define the unit step function u(n) conditionally
x1 = np.zeros_like(n)  # Initialize with zeros
x1[n >= 10] = 1       # Set values to 1 for n >= 0
```

```python
plt.stem(n, x1)
plt.title("x(n) = u(n)")
plt.xlabel("n")
plt.ylabel("x(n)")
plt.grid(True)
plt.axis()
plt.show()
```

```python
#Unit step seqquence for x(n) = u(n) - u(n-10)
x2 = x - x1  # Compute u(n) - u(n-10)
```

```python
plt.stem(n, x2)
plt.title("x(n) = u(n) - u(n-10)")
plt.xlabel("n")
plt.ylabel("x(n)")
plt.grid(True)
plt.axis()
plt.show()
```

Complex value


```python
# Parameters
sigma = -0.1  # Exponential decay (real part)
omega_0 = 2 * np.pi * 0.1  # Angular frequency (imaginary part)
n = np.arange(0, 50)  # Range of n values

# Generate complex-valued sequence
x_n = np.exp((sigma + 1j * omega_0) * n)

# Plot the real and imaginary parts
plt.figure(figsize=(12, 6))
```

```python
# Real part
plt.subplot(2, 1, 1)
plt.stem(n, x_n.real)
plt.title("Real Part of $x(n)$")
plt.xlabel("n")
plt.ylabel("Real(x(n))")
plt.grid()

# Imaginary part
plt.subplot(2, 1, 2)
plt.stem(n, x_n.imag)
plt.title("Imaginary Part of $x(n)$")
plt.xlabel("n")
plt.ylabel("Imag(x(n))")
plt.grid()

plt.tight_layout()
plt.show()

# Optional: Plot complex values in the complex plane
plt.figure(figsize=(6, 6))
plt.plot(x_n.real, x_n.imag, 'o-', label="$x(n)$ in Complex Plane")
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.title("Complex Plane Representation of $x(n)$")
plt.xlabel("Real Part")
plt.ylabel("Imaginary Part")
plt.grid()
plt.legend()
plt.axis('equal')
plt.show()
```

Example Generate complex-valued signal:  


```python
# Parameters
sigma = -0.1  # Exponential decay (real part)
omega_0 = 0.3  # Angular frequency (imaginary part)
n = np.arange(-10, 11)  # Range of n values

# Generate complex-valued sequence
x_n = np.exp((sigma + 1j * omega_0) * n)
```

```python
# Compute the magnitude and phase
magnitude = np.abs(x_n)
phase = np.angle(x_n)
```

```python
# Plot the magnitue, phase, real and imaginary parts
plt.figure(figsize=(12, 10))
```

```python
#Magnitude
plt.stem(n, magnitude)
plt.title("Magnitude of $x(n)$")
plt.xlabel("n")
plt.ylabel("|x(n)|")
plt.grid()
plt.axis()
plt.show()
```

```python
#Phase
plt.stem(n, phase)
plt.title("Phase of $x(n)$")
plt.xlabel("n")
plt.ylabel("x(n)")
plt.grid()
plt.show()
```

```python
#Real part
plt.stem(n, x_n.real)
plt.title("Real Part of $x(n)$")
plt.xlabel("n")
plt.ylabel("Real(x(n))")
plt.grid()
plt.show()
```

```python
#Imaginary part
plt.stem(n, x_n.imag)
plt.title("Imaginary Part of $x(n)$")
plt.xlabel("n")
plt.ylabel("Imag(x(n))")
plt.grid()
plt.show()
```
