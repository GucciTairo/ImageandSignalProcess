```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import dimpulse, dlti, dlsim
```

```python
#Define common usage:
N = 100 #100 samples using to gather the data.
n = np.arange(0, N)
```

```python
# Define the coefficients of the system
# Coefficients of the denominator (feedback terms)
# Coefficients of the numerator (input terms)
b = [2]
a = [1, -1.7, 0.72]
```

```python
# Create the discrete-time linear time-invariant system
system = dlti(b, a)
```

```python
# Compute the impulse response
n_samples = 100  # Number of samples for visualization
t, h = dimpulse(system, n=n_samples)
```

```python
# Extract impulse response values
h = np.squeeze(h)
```

```python
print(h)
```

```python
# Plot the impulse response
plt.stem(t, h)  
plt.title("Impulse Response h(n)")
plt.xlabel("n")
plt.ylabel("h(n)")
plt.grid()
plt.show()
```

```python
# Define the input x(n) = u(n) (unit step function)
x = np.ones_like(n)  # Unit step function
```

```python
# Compute the system's output for x(n) = u(n) (unit step function)
t_out, y_out = dlsim(system, x[:, np.newaxis])  # Adjusted unpacking
y_out = np.squeeze(y_out)
```

```python
# Plot the output y(n)
plt.stem(t_out, y_out)
plt.title("Output y(n) for x(n) = u(n)")
plt.xlabel("n")
plt.ylabel("y(n)")
plt.grid()
plt.show()
```
