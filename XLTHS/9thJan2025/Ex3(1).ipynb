```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the impulse function δ(n)
def delta(n, shift=0):
    #np.where(condition, x, y) returns x if condition is True(1), and y is False(0)
    return np.where(n == shift, 1, 0)
#In this case, if n == shift: The output is 1(the impulse occurs here)
#Else the output is 0
```

```python
# Define the range of n values
N = 40 # Number of samples
n = np.arange(-20, 20)  # From n = -100 to n = 100
```

```python
# Define h(n) = δ(n) - δ(n-10)
h = delta(n) - delta(n, shift=10)
```

```python

# Plot the function
plt.stem(n, h)
plt.title("h(n) = δ(n) - δ(n-10)")
plt.xlabel("n")
plt.ylabel("h(n)")
plt.grid(True)
plt.show()
```

```python
#Question b
# Define the input signal x(n) as the unit step function
x = np.ones(N) #define x(n) = u(n)
```

```python
# Perform the convolution to find the output y(n)
y = np.convolve(x, h, mode='full')[:N]  # Ensure the length is N
```

```python
# Print the output values
print("Output y(n):")
print(y)
```

```python
#Plot the output y(n)
plt.stem(n,y)
plt.title("Output y(n) = x(n) * h(n)")
plt.axis()
plt.grid()
plt.show()
```
