**4 Neighbors have: N4(P) with p coordinate(x,y): The 4-neighbors of a pixel are the pixels that are directly adjacent to it in the horizontal and vertical directions**  
(x-1, y): The pixel above p.  
(x+1, y): The pixel below p.  
(x, y-1): The pixel to the left of p.  
(x, y+1): The pixel to the right of p.  
**4 diagonal neighbors: Diagonal neighbors are the pixels that are adjacent to p diagonally (at the four corners around p).**  
(x-1, y-1): The top-left diagonal neighbor.  
(x+1, y+1): The bottom-right diagonal neighbor.  
(x+1, y-1): The bottom-left diagonal neighbor.  
(x-1, y+1): The top-right diagonal neighbor.  
**N8(p): All 8 surrounding neighbors**  
N8(p) = N4(p) U ND(p)  



4-adjacency: Two pixels are adjacent if one is a 4-neighbor of the other (directly above, below, left, or right).  
8-adjacency: Two pixels are adjacent if one is an 8-neighbor of the other (this includes both 4-neighbors and diagonal neighbors).
