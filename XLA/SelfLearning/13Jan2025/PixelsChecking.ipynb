```python
from PIL import Image
```

```python
# Path to your image
image_path = "/home/tairo/CODETHAYTIEN/XLA/8thJan/ImageFolder/RiceBlast(1).jpg"
```

```python
# Open the image
image = Image.open(image_path)
```

```python
# Convert the image to RGB (if it's not already in RGB mode)
image = image.convert("RGB")
```

```python
# Get image dimensions
width, height = image.size
```

```python
# Print pixel values
print(f"Pixel values of {image_path}:")
for y in range(height):
    for x in range(width):
        pixel = image.getpixel((x, y))  # Get pixel value at (x, y)
        print(f"Pixel at ({x}, {y}): {pixel}")
```

```python
# Create a matrix to store pixel values
pixel_matrix = []

for y in range(height):
    row = []
    for x in range(width):
        pixel = image.getpixel((x, y))  # Get pixel value at (x, y)
        row.append(pixel)  # Append pixel to the current row
    pixel_matrix.append(row)  # Append the row to the matrix
```

```python
# Print the pixel matrix
print("Pixel Matrix:")
for row in pixel_matrix:
    print(row)
```

```python
# Optionally close the image after processing
image.close()
```
