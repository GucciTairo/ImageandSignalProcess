```python
def resize_image(input_path, output_folder, size=(224,224)):
    # Create output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Load and resize image
    with Image.open(input_path) as img:
        # Resize with aspect ratio preservation
        img.thumbnail(size, Image.Resampling.LANCZOS)
        
        # Create new image with exact dimensions
        new_img = Image.new("RGB", size, (255, 255, 255))
        # Paste resized image in center
        offset = ((size[0] - img.size[0]) // 2,
                 (size[1] - img.size[1]) // 2)
        new_img.paste(img, offset)
        
        # Save resized image
        output_path = os.path.join(output_folder, "resized_DogCat.jpeg")
        new_img.save(output_path, "JPEG")
        print(f"Resized image saved to: {output_path}")
```
