I is a grayscale image. Quantize the intensity of I using 4 bits. Show the results. Generalize with k bits(k = 8).

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def quantize_image(image_path, k_bits):
    # Read grayscale image
    img = Image.open(image_path).convert('L')
    img_array = np.array(img)
    
    # Calculate quantization levels and step
    levels = 2**k_bits
    step = 256 / levels
    
    # Quantize image
    quantized = np.floor(img_array / step) * step
    
    return Image.fromarray(quantized.astype(np.uint8))
```

```python
# Load and process image
image_path = "/home/tairo/CODETHAYTIEN/XLA/ImageUseToTest/Gray.jpeg"
original = Image.open(image_path).convert('L')
```

```python
# Create 4-bit quantization
quantized_4bit = quantize_image(image_path, k_bits=4)
```
