```python
def resize_image(input_path, output_folder, upscale_factor=2, downscale_factor=3):
    # Ensure output folder exists
    os.makedirs(output_folder, exist_ok=True)
    # Read the image
    image = cv2.imread(input_path)
    # Check if image was loaded
    if image is None:
        print(f"Unable to load the image from path: {input_path}")
        return
    # Get the filename without extension
    filename = os.path.splitext(os.path.basename(input_path))[0]
    # Downscale the image
    downscale_path = os.path.join(output_folder, f"{filename}_decreasedby3.jpeg") #save the output image in this folder
    try:
        # Use 1 / downscale_factor for scaling
        #Use cv2.resize function from opencv to downscale the images
        decreased_res = cv2.resize(
            image, None, fx=1 / downscale_factor, fy=1 / downscale_factor, interpolation=cv2.INTER_AREA
        )
        cv2.imwrite(downscale_path, decreased_res)
        print(f"Decreased resolution image saved at: {downscale_path}")
    except Exception as e:
        print(f"Error while decreasing resolution: {e}")
```
